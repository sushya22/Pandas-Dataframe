# DataFrame Of Pandas

In pandas library there are 2 data-structures namely.
1.Series datastructure: 1-D labeled array.
2.DataFrame datastructure: 2-D labeled array.

```python
dict = {
    "name":['sushant','virat','rohit','rahul'],
    "score":[80,76,56,43],
    "rank":[4,3,8,6]
}
```

```python
import pandas as pd
df=pd.DataFrame(data=dict)
```
